```python
# Крок 1: Завантаження датасету та виведення перших 10 зображень
import tensorflow as tf
from tensorflow.keras import datasets
import matplotlib.pyplot as plt

# Завантаження CIFAR-10
(x_train, y_train), (x_test, y_test) = datasets.cifar10.load_data()

# Виведення перших 10 зображень
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(x_train[i])
    plt.axis('off')
plt.show()

# Крок 2: Підготовка даних
# Нормалізація пікселів
x_train, x_test = x_train / 255.0, x_test / 255.0

# Перетворення міток на категорії
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)

# Крок 3: Створення моделі
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(10, activation='softmax')
])

# Виведення архітектури моделі
model.summary()

# Крок 4: Налаштування моделі та додавання кол-беків
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Колбеки: рання зупинка та зменшення швидкості навчання
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=0.0001)

# Налаштування оптимізатора, втрат та метрики
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Навчання моделі
history = model.fit(x_train, y_train, epochs=20, batch_size=64,
                    validation_data=(x_test, y_test),
                    callbacks=[early_stopping, reduce_lr])

# Крок 5: Побудова графіків функції втрат та точності
plt.figure(figsize=(12, 4))

# Графік функції втрат
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Графік точності
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.show()

# Крок 6: Класифікаційний звіт
from sklearn.metrics import classification_report

# Прогнозування на тестових даних
y_pred = model.predict(x_test)
y_pred_classes = y_pred.argmax(axis=1)
y_true = y_test.argmax(axis=1)

# Виведення класифікаційного звіту
print(classification_report(y_true, y_pred_classes))

# Крок 7: Візуалізація розпізнавання
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(x_test[i])
    plt.title(f"True: {y_true[i]}, Pred: {y_pred_classes[i]}")
    plt.axis('off')
plt.show()
```

Висновок:Архітектура моделі: У першому завданні було розроблено модель з використанням згорткових шарів для класифікації зображень з датасету CIFAR-10. Було успішно використано кілька згорткових шарів для витягнення важливих ознак зображень. Для покращення продуктивності моделі також було застосовано шар нормалізації, шари pooling та dropout для запобігання перенавчанню.

Підготовка даних: Ми виконали підготовку даних, зокрема нормалізацію пікселів зображень для прискорення процесу навчання та покращення збіжності. Було використано функцію попередньої обробки preprocess_input.

Колбеки: Для ефективного навчання моделі було застосовано callback-и — ранню зупинку та зменшення швидкості навчання при погіршенні результатів на валідації.

Навчання та валідація: Протягом навчання моделі спостерігалося покращення якості класифікації, що підтверджується зменшенням функції втрат і зростанням точності як на навчальних, так і на тестових даних.

Графічне відображення: Було побудовано графіки функції втрат і точності, що допомогло наочно побачити процес навчання та якість моделі.

Класифікаційний звіт: Після навчання було створено класифікаційний звіт, який показав точність моделі для кожного класу.
